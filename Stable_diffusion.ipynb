```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
def get_completion_sd(prompt):
  negative_prompt="""
  bad lighting, bad colors, wrong pallete, wrong picture, bad backgroud, simple background, bad proportions, wrong anatomy,
  duplicate, blurry, low quality, bad quality, lowest quality, watermark
  """

  return pipe(prompt, negative_prompt=negative_prompt).images[0]
```

```python
prompt = "a photo of an astronaut on jupiter with an alien standing"
image = pipe(prompt).images[0]

image.save("./astronaut_2.png")
```

```python
import gradio as gr
```

```python
def get_completion(prompt):
  negative_prompt="""
  bad lighting, bad colors, wrong pallete, wrong picture, bad backgroud, simple background, bad proportions, wrong anatomy,
  duplicate, blurry, low quality, bad quality, lowest quality, watermark, artificial look, artificial colors, vague picture,
  less digits, less pixels, unnatural looks, blurred, low lighting, wrong lighting, incorrect, raciest, too many reflections, incomplete features,
  incomplete faces, incomplete body
  """

  return pipe(prompt, negative_prompt=negative_prompt).images[0]
```

```python
text_to_image_app = gr.Interface(
    fn=get_completion,
    inputs=[gr.Textbox(label="Enter the description of the image to see magic")],
    outputs=[gr.Image(label="Result")],
    title="Text to Image magic",
    description="Generating images from text using Stable Diffusion",
    allow_flagging="never",
    examples=["astronaut on mars","Cargo ship in rough seas"]

)
text_to_image_app.launch()
```
