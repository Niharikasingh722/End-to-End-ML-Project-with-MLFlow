```python
import pandas as pd
```

```python
messages=pd.read_csv("/content/SMSSpamCollection.csv", sep="\t",
names=['label','message'])
```

```python
messages.head()
```

```python
messages.shape
```

data preprocessing: tokenization, remove stop words, special symbols
convert text to lower,
stemming , lemmatization


```python
import re
import nltk
nltk.download("stopwords")
```

Stemming

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(len(messages)):
  review=re.sub('[^a-zA-Z]'," ",messages['message'][i])
  review=review.lower()
  review=review.split(" ")
  review=[ps.stem(word) for word in review if word not in stopwords.words('english')]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000, binary=True)
X=cv.fit_transform(corpus)
```

```python
X=X.toarray()
```

```python
len(X)
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=.30,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
NBModel=MultinomialNB()
```

```python
NBModel=NBModel.fit(train_X,train_y)
```

```python
y_pred=NBModel.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report
```

```python
score=accuracy_score(test_y,y_pred)
score
```

```python
print(classification_report(test_y, y_pred))
```

lemmatize

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(messages)):
  review=re.sub("[^a-zA-Z]"," ",messages['message'][i])
  review=review.lower()
  review=review.split()

  review=[lemma.lemmatize(word) for word in review if word not in stopwords.words('english')]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
  sent_token=sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
model.wv.index_to_key #vocabulary
```

```python
model.wv[0].shape #vector size of each vector
```

```python
model.wv[0]
```

```python
import numpy as np
```

```python
def avgWord2Vec(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
  X.append(avgWord2Vec(words[i]))
```
